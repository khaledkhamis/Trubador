```python
import pandas as pd
import numpy as np
import re
from datetime import datetime
import time
```

```python
df = pd.read_csv('axs_events_2018-2-12_pandas.csv', encoding = 'ansi')
```

```python
df.head(10)
```

```python
#Reorder columns

cols = df.columns.tolist()
cols = cols[4:] + cols[1:4] + cols[0:1]
df = df[cols]
df.head()
```

```python
#Make a date column

df.loc[:,'year'] = pd.Series(2018, index = df.index)
df['month'] = df['month'].map({'JAN': 1, 'FEB': 2, 'MAR': 3, 'APR': 4, 'MAY': 5, 'JUN': 6, 'JUL': 7, 'AUG': 8, 'SEP': 9, 'OCT': 10, 'NOV': 11, 'DEC': 12})
df.loc[:, 'date'] = pd.to_datetime(df.year*10000+df.month*100+df.day,format='%Y%m%d')

# retired
# months = testdf.month.unique()
# months = months.tolist()
```

```python
#Weekend column

df['weekend'] = ((df['weekday'] == 'FRI') | (df['weekday'] == 'SAT') | (df['weekday'] == 'SUN'))
```

```python
#Split performer names

foo = lambda x: pd.Series([i for i in reversed(re.split('with | *: | and', x))])
splitter = df['performer'].apply(foo)
df['performer'] = splitter[0]
df['performer2'] = splitter[1]
df['performer3'] = splitter[2]
df = df.replace(np.nan, '', regex=True)
```

```python
#Clean up venue name

df['venue'].replace(regex=True,inplace=True,to_replace=r',',value=r'')
```

```python
#Remove all duplicate observations

df = df.drop_duplicates()
```

```python
#Clean up time column

splitter2 = df['time'].str.split('- ').to_frame()
for i in splitter2['time']:
    del i[0]
splitter2['time'] = splitter2['time'].apply(lambda x: ''.join(x))
splitter2['time'].replace(regex=True,inplace=True,to_replace=r' ',value=r'')
for i in splitter2:
    splitter2['mytime'] = pd.to_datetime(splitter2['time'])
splitter2['performer'] = df['performer']
df = pd.merge(df, splitter2, on = 'performer', how='left')
df = df.drop(['time_x', 'time_y'], axis=1)
```

```python
#Clean up price

foo = lambda x: pd.Series([i for i in list(re.split('Price: | - |Starting from: ', x))])
splitter3 = df['price'].apply(foo)
splitter3.columns = ['trash', 'price', 'premium_price']
df['price'] = splitter3['price']
df['premium_price'] = splitter3['premium_price']
```

```python
#Reorder columns

cols = df.columns.tolist()
cols
cols = cols[0:4] + cols[14:] + cols[13:14] + cols[6:7] + cols[9:13] + cols[7:8] + cols[4:6] + cols[8:9]
df = df[cols]
```

### Ticket Price by Weekday

```python
from re import sub
from decimal import Decimal

df = df.replace(np.nan, '', regex=True)
df['price'] = df['price'].str.replace('$', '')
#df = df.drop([25])
```

```python
Weekend = df.loc[df.weekend==True]['price']
Weekday = df.loc[df.weekend==False]['price']

getbins = lambda t:{'start':t.min(),'end':t.max(), 'size':10}

trace0 = Histogram(x=Weekend,
                   name='Weekend',
                   xbins=getbins(Weekend))

trace1 = Histogram(x=Weekday,
                   name='Weekday',
                   xbins=getbins(Weekend))

data = [trace0, trace1]

layout = Layout(title="Histogram of Weekend and Weekday Ticket Prices",
               bargroupgap=0.1)
fig = Figure(data=data, layout=layout)

plotly.offline.iplot(fig, show_link=False, image_width=600, image_height=400)
```
